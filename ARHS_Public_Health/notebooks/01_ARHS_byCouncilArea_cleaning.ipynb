#### Data Cleaning - Alcohol-Related Hospital Statistics by Council Area (Scotland)

##### Objective
Prepare a **clean, analysis-ready dataset** for downstream Python analysis and Power BI reporting.

**Primary outcome variable:**
- easr_patients (European age-standardised rate per 100,000)

```python
#Import python libraries

import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Project paths

DATA_DIR = Path('../data')
RAW_DIR = DATA_DIR / 'raw'
PROCESSED_DIR = DATA_DIR / 'processed'
FIGURES_DIR = Path('../figures')
```

```python
#Load raw data - ARHS dataset by council area

if Path('/Volumes').exists():
    RAW_DIR = Path('/Volumes/group_project_wfs2025/default/arhs')
else:
    RAW_DIR = Path('../data/raw')

arhs_council_df = pd.read_csv(RAW_DIR/'ARHS_byCouncilArea.csv')

arhs_council_df
```

```python
#Summary info
print(arhs_council_df.info())

#Duplicate values
print(arhs_council_df.duplicated().sum())

#Null values
print(arhs_council_df.isna().sum())
```

```python
#Drop columns - not necessary for further analysis
#Drop the following columns - NumberOfStaysQF, NumberOfPatientsQF, NumberOfNewPatientsQF, EASRNewPatientsQF, CAQF, ICD10QF, ICD10

arhs_council_df.drop(
    ['NumberOfStaysQF', 'NumberOfPatientsQF', 'NumberOfNewPatientsQF', 'EASRNewPatientsQF', 'CAQF', 'ICD10QF', 'ICD10'],
    axis=1,
    inplace = True
)

arhs_council_df
```

```python
#Uniqe values in each columns
print(arhs_council_df.nunique())

print('\nUniqe values of Financial Year: ')
print(arhs_council_df['FinancialYear'].unique())

print('\nUniques values of Council Areas: ')
print(arhs_council_df['CA'].unique())

print('\nUnique values of SMRType: ')
print(arhs_council_df['SMRType'].unique())

print('\nUnique values of Condition: ')
print(arhs_council_df['Condition'].unique())
```

```python
#FinancialYear column

#Rename FinancalYear column to FinancialYear_label
arhs_council_df.rename(
    columns = {'FinancialYear' : 'FinancialYear_label'},
    inplace = True
)

#Convert object datatype to str

arhs_council_df['FinancialYear_label'] = arhs_council_df['FinancialYear_label'].astype('string')

#Extarct first four character from FinancialYear_label as FinancialYear

arhs_council_df['FinancialYear'] = (
    arhs_council_df['FinancialYear_label']
    .str.extract(r'(\d{4})', expand = False)
    .astype('int')
)

#Validate change

print(arhs_council_df.info())
```

```python
#Harmonise column names

arhs_council_df.rename(
    columns={
        'Condition': 'condition',
        'SMRType': 'smr_type_code',
        'FinancialYear_label': 'financial_year_label',
        'CA': 'council_code',
        'EASRStays': 'easr_stays',
        'EASRPatients': 'easr_patients',
        'EASRNewPatients': 'easr_new_patients',
        'NumberOfStays': 'number_of_stays',
        'NumberOfPatients': 'number_of_patients',
        'NumberOfNewPatients': 'number_of_new_patients',
        'AverageNumberOfStaysPerPatient': 'average_number_of_stays_per_patient',
        'FinancialYear': 'financial_year'
    },
    inplace=True
)
```

```python
#Create dictionary for mapping council areas

COUNCIL_CODE_MAP = {
    'S92000003' :'Scotland',
    'S12000005' : 'Clackmannanshire',
    'S12000006' : 'Dumfries and Galloway',
    'S12000008' : 'East Ayrshire',
    'S12000010' : 'East Lothian',
    'S12000011' : 'East Renfrewshire',
    'S12000013' : 'Na h-Eileanan Siar',
    'S12000014' : 'Falkirk',
    'S12000017' : 'Highland',
    'S12000018' : 'Inverclyde',
    'S12000019' : 'Midlothian',
    'S12000020' : 'Moray',
    'S12000021' : 'North Ayrshire',
    'S12000023' : 'Orkney Islands',
    'S12000026' : 'Scottish Borders',
    'S12000027' : 'Shetland Islands',
    'S12000028' : 'South Ayrshire',
    'S12000029' : 'South Lanarkshire',
    'S12000030' : 'Stirling',
    'S12000033' : 'Aberdeen City',
    'S12000034' : 'Aberdeenshire',
    'S12000035' : 'Argyll and Bute',
    'S12000036' : 'City of Edinburgh',
    'S12000038' : 'Renfrewshire',
    'S12000039' : 'West Dunbartonshire',
    'S12000040' : 'West Lothian',
    'S12000041' : 'Angus',
    'S12000042' : 'Dundee City',
    'S12000045' : 'East Dunbartonshire',
    'S12000047' : 'Fife',
    'S12000048' : 'Perth and Kinross',
    'S12000049' : 'Glasgow City',
    'S12000050' : 'North Lanarkshire', 
}

SMR_TYPE_MAP = {
    'SMR01' : 'Non-psychiatric',
    'SMR04' : 'Psychiatric',
    'Combined' : 'Combined'
}

#Map council areas and SMR types

arhs_council_df['council_area'] = arhs_council_df['council_code'].map(COUNCIL_CODE_MAP)
arhs_council_df['smr_type'] = arhs_council_df['smr_type_code'].map(SMR_TYPE_MAP)
arhs_council_df
```

```python
#New variable
string_cols = [
    '_id', 'financial_year_label', 'condition',
    'council_code', 'council_area', 'smr_type'
]

#Ensure dataypes in each columns
for col in string_cols:
    arhs_council_df[col] = arhs_council_df[col].astype('string')


#Validate clean dataset
print('\nSummary info of clean dataset: ')
print(arhs_council_df.info())

#Unique values of clean dataset
print('\nUnique values of clean dataset: ')
print(arhs_council_df.nunique())

```

```python
#Validation check

assert arhs_council_df['financial_year'].notna().all()
assert arhs_council_df['council_area'].notna().all()
```

```python
#Export clean dataset to csv

if Path('/Volumes').exists():
    PROCESSED_DIR = Path('/Volumes/group_project_wfs2025/default/arhs')
else:
    PROCESSED_DIR = Path('../data/processed')

arhs_council_df.to_csv(
    PROCESSED_DIR / 'ARHS_byCouncilArea_clean.csv',
    index=False
)
```
